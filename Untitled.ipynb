```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 
from scipy.signal import argrelextrema
from scipy.interpolate import CubicSpline

from astropy.io import fits as pyfits

matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})
```

```python
def resid(t, f, anchor=True):
    '''
    Outline:
    * do 1-point derivative to find local Max's and Min's
        * this is point of possible improvement
    * do cubic splines between Max's and Min's
    * find running mean between Max and Min
    * return residial
    '''
    maxm = argrelextrema(f, np.greater)
    minm = argrelextrema(f, np.less)

    if anchor:
        maxm = np.append([0], maxm)
        minm = np.append([0], minm)
        maxm = np.append(maxm,f.size-1)
        minm = np.append(minm,f.size-1)
        
        maxm = np.unique(maxm)
        minm = np.unique(minm)

#     if (len(maxm) < 2) | (len(minm) < 2):
#         print('Not Enough Extrema!')
#         return f
        
    maxCS = CubicSpline(t[maxm], f[maxm])
    maxi = maxCS(t, extrapolate=None)

    minCS = CubicSpline(t[minm], f[minm])
    mini = minCS(t, extrapolate=None)

    meani = (maxi + mini) / 2.0
    
    return f - meani
 

def IMF(t, f, Nmax=500, SDmax=0.0001, Zmax=5):
    ftmp = f
    i = 0
    Zok = 0
    
    while i<Nmax:
        r1 = resid(t, ftmp)
    
        # does the residual satisfy the 2 criteria?
        # 1) Nextrema = Nzero_crossings
        # 2) StdDev is lower than some threshold
        SD = np.sum(np.power(np.abs(ftmp - r1), 2) / np.power(ftmp, 2))

        # compute Nextrema
        maxm = argrelextrema(r1, np.greater)
        minm = argrelextrema(r1, np.less)

        # from http://stackoverflow.com/a/30281079
        Nzero = ((r1[:-1] * r1[1:]) < 0).sum()

        if np.abs(Nzero - (np.size(maxm)+np.size(minm))) < 2:
            Zok = Zok + 1
        if SD <= SDmax:
            i=Nmax
        if Zok >= Zmax:
            i=Nmax
        i=i+1
        ftmp = r1
        
    return r1

def EMD10(t, f, display=False, N=10):
    ftmp = f.copy()
    output = np.zeros((N, f.size))
    
    for i in range(N):
        mode_i = IMF(t, ftmp)
        ftmp = ftmp - mode_i
        
        output[i,:] = mode_i
        
        if display:
            plt.plot(t, mode_i + i * np.std(f))
            
    return output
```

```python
time = f[1].data['TIME']
fraw = np.sum(f[1].data['FLUX'],axis=(1,2))

fit = np.polyfit(time, fraw, 4)
pp = np.poly1d(fit)

plt.figure(figsize=(11,6))
plt.plot(time, fraw - pp(time))
plt.ylim(-5000, 5000)

x = np.where((fraw - pp(time) > -2000))
plt.plot(time[x], fraw[x] - pp(time[x]))
```
